```python
import pandas as pd

df = pd.read_csv("../data/raw/sleep_data.csv", on_bad_lines="skip")
```

#### Remove the naps column 

```python
df = df.drop(["naps"], axis=1)
```

#### Remove "+0000" from the start and stop columns

```python
def strip_time(time):
    time = time.split(" ")[1].split("+")[0]
    return time

df["start"] = df["start"].astype(str).apply(strip_time)
df["stop"] = df["stop"].astype(str).apply(strip_time)
```

#### Turn the start and stop times to minutes after 00:00

```python
def time_to_minutes(time):
    hours = int(time.split(":")[0])
    minutes = int(time.split(":")[1])
    time = (hours * 60) + minutes
    return time

df["start"] = df["start"].apply(time_to_minutes)
df["stop"] = df["stop"].apply(time_to_minutes)
```

#### Write the cleaned data to a csv file 

```python
df.to_csv("../data/cleaned/sleep_data_cleaned.csv", index=False)
```
