```python
import pandas as pd 
import matplotlib.pyplot as plt

df = pd.read_csv("../data/cleaned/combined_data_cleaned.csv")

print(df)
```

```python
x = df["date"]
y1 = df["totalSleepTime"]
y2 = df["deepSleepTime"]
y3 = df["shallowSleepTime"]
y4 = df["REMTime"]

fig, axs = plt.subplots(2, 2)

axs[0, 0].plot(x, y1)
axs[0, 0].set_title("Total Sleep Time")
axs[0, 1].plot(x, y2)
axs[0, 1].set_title("Deep Sleep Time")
axs[1, 0].plot(x, y3)
axs[1, 0].set_title("Shallow Sleep Time")
axs[1, 1].plot(x, y2)
axs[1, 1].set_title("REM Sleep Time")
```
