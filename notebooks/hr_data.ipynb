```python
import pandas as pd

df = pd.read_csv("../data/raw/hr_data.csv", on_bad_lines="skip")
```

#### Find the average heart rate for each day

```python
total_hr = df["heartRate"].groupby(df["date"])
cleaned_df = total_hr.mean().astype(int)
```

#### Write the cleaned data to a csv file

```python
cleaned_df.to_csv("../data/cleaned/hr_data_cleaned.csv")
```
